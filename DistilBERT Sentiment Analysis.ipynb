```python
from transformers import pipeline

model_name = 'distilbert-base-uncased-finetuned-sst-2-english'
sentiment_analysis = pipeline("sentiment-analysis", model=model_name)
```

```python
import pandas as pd

# read test.EN.csv into a dataframe
df_test = pd.read_csv('https://raw.githubusercontent.com/iabufarha/iSarcasmEval/main/test/task_A_En_test.csv', index_col=False)
```

```python
test_features = df_test.text.values
test_labels = df_test.sarcastic.values
```

```python
y_hat = []
for text in test_features:
    y_hat.append(sentiment_analysis(text)[0]['label'] == 'NEGATIVE')
```

```python
from sklearn.metrics import f1_score, accuracy_score
print("F1 score of baseline sentiment analysis model is", f1_score(test_labels, y_hat))
```
