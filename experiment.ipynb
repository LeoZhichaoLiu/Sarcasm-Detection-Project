```python
import torch
import torch.nn as nn
from torch.nn import functional as F
import math

class Fusion1(nn.Module):
  def __init__(self,len_data,dim_in,dim_k):
    super(Fusion1,self).__init__()
    self.MHSA=MultiHeadSelfAttention(len_data,dim_in,dim_k,dim_k)
    self.FFN=FFN(dim_k)
    self.simple_attention=SimpleAttention(dim_k)
    self.layernorm=nn.LayerNorm(dim_k)
    self.dropout=nn.Dropout(p=0.1)
    self.fc_dim1=dim_k#sequnece length
    self.fc_dim2=32
    self.fc_dim3=16
    self.fc_dim4 = 1
    self.fc_dim5 = 1
    self.fc1 = nn.Linear(self.fc_dim1, self.fc_dim2)
    self.fc2 = nn.Linear(self.fc_dim2, self.fc_dim3)
    self.fc3 = nn.Linear(self.fc_dim3, self.fc_dim4)
    self.fc4 = nn.Linear(self.fc_dim4, self.fc_dim5)
    self.output_layer=nn.Sigmoid()
    self.bn1 = nn.BatchNorm1d(num_features=self.fc_dim2, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn2 = nn.BatchNorm1d(num_features=self.fc_dim3, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn3 = nn.BatchNorm1d(num_features=self.fc_dim4, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)

  def forward(self, x):
    x_mhsa=self.MHSA(x)
    x=self.layernorm(x+x_mhsa)
    x=self.dropout(x)
    x=self.FFN(x)
    #x=torch.reshape(x,(x.shape[0],-1))
    x=self.simple_attention(x)
    x=self.fc1(x)
    x=F.relu(x)
    x=self.bn1(x)
    x=self.dropout(x)
    x=self.fc2(x)
    x=F.relu(x)
    x=self.bn2(x)
    x=self.dropout(x)
    x=self.fc3(x)
    x=self.output_layer(x)
    return x

class Fusion3(nn.Module):
  def __init__(self,len_data,dim_in,dim_k):
    super(Fusion3,self).__init__()
    self.encoder1=EncoderLayers(len_data,dim_in,dim_k)
    self.encoder2=EncoderLayers(len_data,dim_in,dim_k)
    self.MHSA=MultiHeadSelfAttention(len_data,dim_in,dim_k,dim_k)
    self.FFN=FFN(dim_k)
    self.simple_attention=SimpleAttention(dim_k)
    self.layernorm=nn.LayerNorm(dim_k)
    self.dropout=nn.Dropout(p=0.5)
    self.fc_dim1=dim_k*2#sequnece length
    self.fc_dim2=32
    self.fc_dim3=16
    self.fc_dim4 = 1
    self.fc_dim5 = 1
    self.fc1 = nn.Linear(self.fc_dim1, self.fc_dim2)
    self.fc2 = nn.Linear(self.fc_dim2, self.fc_dim3)
    self.fc3 = nn.Linear(self.fc_dim3, self.fc_dim4)
    self.fc4 = nn.Linear(self.fc_dim4, self.fc_dim5)
    self.output_layer=nn.Sigmoid()
    self.bn1 = nn.BatchNorm1d(num_features=self.fc_dim2, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn2 = nn.BatchNorm1d(num_features=self.fc_dim3, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn3 = nn.BatchNorm1d(num_features=self.fc_dim4, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)

  def forward(self, x1,x2):
    x1=self.encoder1(x1)
    x2=self.encoder2(x2)
    x=torch.cat([x1,x2],dim=1)
    x=self.fc1(x)
    x=F.relu(x)
    x=self.bn1(x)
    x=self.dropout(x)
    x=self.fc2(x)
    x=F.relu(x)
    x=self.bn2(x)
    x=self.dropout(x)
    x=self.fc3(x)
    x=self.output_layer(x)
    return x

class Fusion2(nn.Module):
  def __init__(self,len_data,dim_in,dim_k):
    super(Fusion2,self).__init__()
    self.MHSA=CrossMultiHeadSelfAttention(len_data,dim_in,dim_k,dim_k)
    self.FFN=FFN(dim_k)
    self.simple_attention=SimpleAttention(dim_k)
    self.layernorm=nn.LayerNorm(dim_k)
    self.dropout=nn.Dropout(p=0.5)
    self.fc_dim1=dim_k#sequnece length
    self.fc_dim2=32
    self.fc_dim3=16
    self.fc_dim4 = 1
    self.fc_dim5 = 1
    self.fc1 = nn.Linear(self.fc_dim1, self.fc_dim2)
    self.fc2 = nn.Linear(self.fc_dim2, self.fc_dim3)
    self.fc3 = nn.Linear(self.fc_dim3, self.fc_dim4)
    self.fc4 = nn.Linear(self.fc_dim4, self.fc_dim5)
    self.output_layer=nn.Sigmoid()
    self.bn1 = nn.BatchNorm1d(num_features=self.fc_dim2, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn2 = nn.BatchNorm1d(num_features=self.fc_dim3, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)
    self.bn3 = nn.BatchNorm1d(num_features=self.fc_dim4, eps=1e-05, momentum=0.1, affine=True,
                              track_running_stats=True)

  def forward(self, x1,x2):
    x_mhsa=self.MHSA(x1,x2)
    x=self.layernorm(x1+x_mhsa)
    x=self.dropout(x)
    x=self.FFN(x)
    #x=torch.reshape(x,(x.shape[0],-1))
    x=self.simple_attention(x)
    x=self.fc1(x)
    x=F.relu(x)
    x=self.bn1(x)
    x=self.dropout(x)
    x=self.fc2(x)
    x=F.relu(x)
    x=self.bn2(x)
    x=self.dropout(x)
    x=self.fc3(x)
    x=self.output_layer(x)
    return x

class EncoderLayers(nn.Module):
    def __init__(self,len_data,dim_in,dim_k):
      super(EncoderLayers,self).__init__()
      self.MHSA=MultiHeadSelfAttention(len_data,dim_in,dim_k,dim_k)
      self.FFN=FFN(dim_k)
      self.simple_attention=SimpleAttention(dim_k)
      self.layernorm=nn.LayerNorm(dim_k)
      self.dropout=nn.Dropout(p=0.5)


    def forward(self, x):
      x_mhsa=self.MHSA(x)
      x=self.layernorm(x+x_mhsa)
      x=self.dropout(x)
      x=self.FFN(x)
      x=self.simple_attention(x)
      return x

class MultiHeadSelfAttention(nn.Module):
    dim_in: int  # input dimension
    dim_k: int   # key and query dimension
    dim_v: int   # value dimension
    num_heads: int  # number of heads, for each head, dim_* = dim_* // num_heads

    def __init__(self, length,dim_in, dim_k, dim_v, num_heads=2):
        super(MultiHeadSelfAttention, self).__init__()
        assert dim_k % num_heads == 0 and dim_v % num_heads == 0, "dim_k and dim_v must be multiple of num_heads"
        self.dim_in = dim_in
        self.dim_k = dim_k
        self.dim_v = dim_v
        self.num_heads = num_heads
        self.linear_q = nn.Linear(dim_in, dim_k, bias=False)
        self.linear_k = nn.Linear(dim_in, dim_k, bias=False)
        self.linear_v = nn.Linear(dim_in, dim_v, bias=False)
        self._norm_fact = 1 / math.sqrt(dim_k // num_heads)
        self.layer_norm=nn.LayerNorm(dim_k)
        self.dropout=nn.Dropout(p=0.1)


    def forward(self, x):
        # x: tensor of shape (batch, n, dim_in)
        batch, n, dim_in = x.shape
        assert dim_in == self.dim_in

        nh = self.num_heads
        dk = self.dim_k // nh  # dim_k of each head
        dv = self.dim_v // nh  # dim_v of each head

        q = self.linear_q(x).reshape(batch, n, nh, dk).transpose(1, 2)  # (batch, nh, n, dk)
        k = self.linear_k(x).reshape(batch, n, nh, dk).transpose(1, 2)  # (batch, nh, n, dk)
        v = self.linear_v(x).reshape(batch, n, nh, dv).transpose(1, 2)  # (batch, nh, n, dv)

        #attn_mask = attn_mask.unsqueeze(1).repeat(1, self.num_heads, 1, 1)

        dist = torch.matmul(q, k.transpose(2, 3)) * self._norm_fact  # batch, nh, n, n
        #dist=dist.masked_fill_(attn_mask, -1e9)
        dist = torch.softmax(dist, dim=-1)  # batch, nh, n, n
        dist=self.dropout(dist)
        att = torch.matmul(dist, v)  # batch, nh, n, dv
        att = att.transpose(1, 2).reshape(batch, n, self.dim_v)  # batch, n, dim_v
        #att_mask=att.eq(0)
        #att=att.masked_fill_(att_mask,-1e9)

        return att

    def get_attention_padding_mask(self, q):
        attn_pad_mask = q[:,:,0].eq(0).unsqueeze(1).repeat(1, q.size(1), 1)
        # |attn_pad_mask| : (batch_size, q_len, k_len)

        return attn_pad_mask

class CrossMultiHeadSelfAttention(nn.Module):
    dim_in: int  # input dimension
    dim_k: int   # key and query dimension
    dim_v: int   # value dimension
    num_heads: int  # number of heads, for each head, dim_* = dim_* // num_heads

    def __init__(self, length,dim_in, dim_k, dim_v, num_heads=2):
        super(CrossMultiHeadSelfAttention, self).__init__()
        assert dim_k % num_heads == 0 and dim_v % num_heads == 0, "dim_k and dim_v must be multiple of num_heads"
        self.dim_in = dim_in
        self.dim_k = dim_k
        self.dim_v = dim_v
        self.num_heads = num_heads
        self.linear_q = nn.Linear(dim_in, dim_k, bias=False)
        self.linear_k = nn.Linear(dim_in, dim_k, bias=False)
        self.linear_v = nn.Linear(dim_in, dim_v, bias=False)
        self._norm_fact = 1 / math.sqrt(dim_k // num_heads)
        self.layer_norm=nn.LayerNorm(dim_k)
        self.dropout=nn.Dropout(p=0.5)


    def forward(self, x1,x2):
        # x: tensor of shape (batch, n, dim_in)
        batch, n, dim_in = x1.shape
        assert dim_in == self.dim_in

        nh = self.num_heads
        dk = self.dim_k // nh  # dim_k of each head
        dv = self.dim_v // nh  # dim_v of each head

        q = self.linear_q(x2).reshape(batch, n, nh, dk).transpose(1, 2)  # (batch, nh, n, dk)
        k = self.linear_k(x2).reshape(batch, n, nh, dk).transpose(1, 2)  # (batch, nh, n, dk)
        v = self.linear_v(x1).reshape(batch, n, nh, dv).transpose(1, 2)  # (batch, nh, n, dv)

        #attn_mask = attn_mask.unsqueeze(1).repeat(1, self.num_heads, 1, 1)

        dist = torch.matmul(q, k.transpose(2, 3)) * self._norm_fact  # batch, nh, n, n
        #dist=dist.masked_fill_(attn_mask, -1e9)
        dist = torch.softmax(dist, dim=-1)  # batch, nh, n, n
        dist=self.dropout(dist)
        att = torch.matmul(dist, v)  # batch, nh, n, dv
        att = att.transpose(1, 2).reshape(batch, n, self.dim_v)  # batch, n, dim_v
        #att_mask=att.eq(0)
        #att=att.masked_fill_(att_mask,-1e9)

        return att

    def get_attention_padding_mask(self, q):
        attn_pad_mask = q[:,:,0].eq(0).unsqueeze(1).repeat(1, q.size(1), 1)
        # |attn_pad_mask| : (batch_size, q_len, k_len)

        return attn_pad_mask

class FFN(nn.Module):
    #feed foward network
    def __init__(self,dim_v):
        super(FFN, self).__init__()
        self.d_ff=100
        self.w1=nn.Linear(dim_v,self.d_ff)
        self.w2=nn.Linear(self.d_ff,dim_v)
        self.layernorm=nn.LayerNorm(dim_v)
        self.dropout=nn.Dropout(p=0.1)


    def forward(self,x):
        residual=x
        x=self.w2(self.dropout(self.w1(x)))
        x=self.layernorm(x+residual)
        x=self.dropout(x)

        return x

class SimpleAttention(nn.Module):
    def __init__(self, input_dim):
        super(SimpleAttention, self).__init__()
        self.input_dim = input_dim
        self.scalar = nn.Linear(self.input_dim, 1, bias=False)

    def forward(self, M,x=None):
        scale = self.scalar(M)
        #scale=scale.masked_fill_(attn_mask, -1e9)
        alpha = F.softmax(scale, dim=1).permute(0, 2, 1)
        attn_pool = torch.bmm(alpha, M)[:, 0, :]

        return attn_pool

net=Fusion1(50,50,50)
net=Fusion2(50,50,50)
input=torch.rand(32,50,50)
input1=torch.rand(32,50,50)
input2=torch.rand(32,50,50)
print(net(input1,input2).shape)
```

```python
import numpy as np
from torch.utils.data.dataset import Dataset
from torch.utils.data.dataloader import DataLoader
import torch
from tqdm import tqdm
import time
import torch.nn as nn
#from sklearn.metrics import precision_score, recall_score, f1_score
from torchmetrics.classification import BinaryF1Score,BinaryPrecision,BinaryRecall,BinaryAUROC


batchsize=64
lr=0.001
epochs=100
criterion = nn.BCELoss()
#criterion=nn.MSELoss(size_average=None, reduce=None, reduction='mean')
dim=50
net1=Fusion1(50,dim,dim)
net2=Fusion2(50,dim,dim)
net3=Fusion3(50,dim,dim)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

def train_and_test(model,pre=False):
    train_dataset=TextDataset('/content/drive/Shareddrives/CSE256/data/train_data_50d.txt',pre,dim)
    train_data,test_data=torch.utils.data.random_split(train_dataset,[int(len(train_dataset)*0.8),len(train_dataset)-int(len(train_dataset)*0.8)])
    train_loader=DataLoader(dataset=train_data,batch_size=32,shuffle=True)
    test_loader=DataLoader(dataset=test_data,batch_size=32,shuffle=True)
    train_loss_all=[]
    train_acc_all=[]
    test_loss_all=[]
    test_acc_all=[]
    p_train=[]
    r_train=[]
    f_train=[]
    a_train=[]
    p_test=[]
    r_test=[]
    f_test=[]
    a_test=[]
    for epoch in tqdm(range(epochs)):
        print('current epoch:'+str(epoch))
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)
        epoch_loss=0
        train_loss=0
        train_acc=0
        test_acc=0
        count=0
        corr=0
        accuracy=0
        precision_train=0
        recall_train=0
        f1_train=0
        auc_train=0
        precision_test=0
        recall_test=0
        f1_test=0
        auc_test=0
        for batch_data in train_loader:
            #data size:(batch_size,num_words,num_dim)
            #label size:(batch_size)
            #label.to(torch.float32)
            model=model.to(device)
            model.train()
            if pre:
                data,label=batch_data
                x=data.to(device=device,dtype=torch.float32)
                y=model(x)
            else:
                data1,data2,label=batch_data
                x1=data1.to(device=device,dtype=torch.float32)
                x2=data2.to(device=device,dtype=torch.float32)
                y=model(x1,x2)
            label=label.unsqueeze(1)
            label=label.to(device=device)
            optimizer.zero_grad()
            loss = criterion(y, label.float())
            loss=loss.to(device=device,dtype=torch.float32)
            loss.backward()
            optimizer.step()
            train_loss+=loss.item()
            pred_binary = (y > 0.5).float()
            correct = (pred_binary.to(device=device) == label.to(device=device)).sum().item()
            ones=torch.ones(label.shape[0],1).to(device)
            zeros=torch.zeros(label.shape[0],1).to(device)

            accuracy += correct / y.size(0)
            #epoch_loss += loss.cpu().detach().numpy().tolist()
            #y = y.cpu().detach().numpy().tolist()
            count+=1
            #true_positive = ((pred_binary == ones) & (label == ones)).sum().item()
            #false_positive = ((pred_binary == ones) & (label == zeros)).sum().item()
            #false_negative = ((pred_binary == zeros) & (label == ones)).sum().item()

            precision_score=BinaryPrecision().to(device)
            recall_score=BinaryRecall().to(device)
            f1_score=BinaryF1Score().to(device)
            auc_score=BinaryAUROC().to(device)

            precision_train += precision_score(y,label)
            recall_train += recall_score(y,label)
            f1_train += f1_score(y,label)
            auc_train+=auc_score(y,label)

        train_acc=accuracy/count
        epoch_loss=epoch_loss/count
        train_loss/=count
        train_acc_all.append(train_acc)
        train_loss_all.append(train_loss)
        p_train.append(precision_train/count)
        r_train.append(recall_train/count)
        f_train.append(f1_train/count)
        a_train.append(auc_train/count)
        print('train mse='+str(train_loss))
        print('train_acc'+str(train_acc))
        print('train_f1'+str(f1_train/count))
        count=0
        corr=0
        test_loss=0
        accuracy=0

        for batch_data in test_loader:
            #label.to(torch.float32)
            model.eval()
            with torch.no_grad():
                if pre:
                    data,label=batch_data
                    x=data.to(device=device,dtype=torch.float32)
                    y=model(x)
                else:
                    data1,data2,label=batch_data
                    x1=data1.to(device=device,dtype=torch.float32)
                    x2=data2.to(device=device,dtype=torch.float32)
                    y=model(x1,x2)
                label=label.unsqueeze(1)
                y = y.cpu().detach()
                count+=1
                test_loss+=criterion(y, label.float()).item()
                pred_binary = (y > 0.5).float()
                correct = (pred_binary.to(device=device) == label.to(device=device)).sum().item()
                ones=torch.ones(label.shape[0],1).to(device)
                zeros=torch.zeros(label.shape[0],1).to(device)
                accuracy += correct / y.size(0)
                #true_positive = ((pred_binary == ones) & (label == ones)).sum().item()
                #false_positive = ((pred_binary == ones) & (label == zeros)).sum().item()
                #false_negative = ((pred_binary == zeros) & (label == ones)).sum().item()

                precision_score=BinaryPrecision()
                recall_score=BinaryRecall()
                f1_score=BinaryF1Score()
                auc_score=BinaryAUROC().to(device)

                precision_test += precision_score(y,label)
                recall_test += recall_score(y,label)
                f1_test += f1_score(y,label)
                auc_test+=auc_score(y,label)

        test_acc=accuracy/count
        test_loss/=count
        test_acc_all.append(test_acc)
        test_loss_all.append(test_loss)
        p_test.append(precision_test/count)
        r_test.append(recall_test/count)
        f_test.append(f1_test/count)
        a_test.append(auc_test/count)
        if pre:
            torch.save({"train_loss":train_loss_all,
                        "train_acc":train_acc_all,
                        "test_loss":test_loss_all,
                        "test_acc":test_acc_all,
                        "test_p":p_test,
                        "test_r":r_test,
                        "test_f1":f_test,
                        "test_auc":a_test,
                        "weights":model.state_dict()},f'/content/drive/Shareddrives/CSE256/train_pre_concatenate_50d.pth')
        else:
            torch.save({"train_loss":train_loss_all,
                        "train_acc":train_acc_all,
                        "test_loss":test_loss_all,
                        "test_acc":test_acc_all,
                        "test_p":p_test,
                        "test_r":r_test,
                        "test_f1":f_test,
                        "test_auc":a_test,
                        "weights":model.state_dict()},f'/content/drive/Shareddrives/CSE256/train_post_concatenate_50d.pth')
        print('test mse='+str(test_loss))
        print('test_acc='+str(test_acc))
        print('test_f1'+str(f1_test/count))
        print('test_auc'+str(auc_test/count))
        print('epoch:'+str(epoch)+' '+'train_mse:'+str(train_loss)+' '+'test_mse:'+str(test_loss))

if __name__=="__main__":
    train_and_test(net3,pre=False)

```
